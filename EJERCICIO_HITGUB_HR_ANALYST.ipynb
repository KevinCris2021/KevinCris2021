```python
func_living = df[(df['Ultima_eva']>=0.7)| (df['Tiempo_Cia']>=4) | (df['Num_proyect']>=5)]
```

```python
func_living
```

```python
correlation2 = func_living.corr()
sns.set(style= 'white' )
mask = np.zeros_like(correlation2,dtype =np.bool)
mask[np.triu_indices_from(mask)] = True

#Grfico

f,ax = plt.subplots(figsize =(13,8))
cmap = sns.diverging_palette(10,220,as_cmap=True)

#Dibujo del heatmap de la mascara

ax = sns.heatmap(correlation2,mask=mask, cmap= cmap, vmax = 5, annot=True, annot_kws={'size':11}, square=True, 
                 xticklabels= True, yticklabels = True, linewidths = 5, cbar_kws ={'shrink':5},ax =ax)

ax.set_title('Correlacion: Por qu los empleados quieren irse?',fontsize =20)
```

Resumen del anlisis de datos exploratorios
- Es una empresa relativamente joven, en promedio, los empleados tienen 3 o 4 aos en la empresa y los empleados de mayor edad estn trabajando 10 aos.
- La mayor diferencia en el salario de los que se quedaron y los que se fueron, se encontr en el departamento de administracin, en los dems departamentos aunque los salarios de los que se quedaron son ms altos en promedio, no es una gran diferencia.
- El nmero de empleados que tuvo un accidente de trabajo es de aproximadamente el 14%, de los cuales solo 169 empleados dejaron la empresa, por lo que no parece tener una correlacin con la salida de los empleados.
- En cinco aos solo el 2% de los empleados fueron promovidos. Es posible que muchos empleados se desmotiven y comiencen a planear irse.
- Los empleados con 7 o ms en la empresa no se fueron. Los empleados con 5 aos tienen ms posibilidades de irse.
- Hay 2 grupos distintos de desempeo de empleados que se fueron. Un grupo con bajo desempeo con 2 proyectos y otros con alto desempeo con 5 o ms proyectos. No es necesario retener a todos los empleados, el enfoque est en mantener a los empleados con un alto rendimiento.
- Los empleados con 4 aos en la empresa tienen el nivel medio de satisfaccin ms bajo de toda la empresa con (0,47).
La satisfaccin cae cuando los empleados estn trabajando en 5 o ms proyectos. Un nmero de 3 o 4 proyectos parece ser ideal independientemente del tiempo dedicado a la empresa.
- Los empleados con 5 o ms proyectos que salieron tambin trabajaron al menos un 20% ms de horas que la media de la empresa.
- El nivel de satisfaccin de los empleados que se fueron se agrupa en totalmente decepcionados, por debajo de la media de satisfaccin y satisfechos.
