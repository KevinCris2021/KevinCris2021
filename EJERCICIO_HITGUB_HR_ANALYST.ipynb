¿Cuándo los empleados se sienten insatisfechos?
En los próximos resultados se evidencia la caída en la satisfacción cuando los empleados están trabajando en 6 o más proyectos.

Veamos por qué los empleados más valiosos tienden a irse.

De los empleados que salieron con alto desempeño, 4 o más años en la empresa y trabajando en 5 o más proyectos tuvieron:

Nivel de satisfacción bajo,
Trabajó más horas,
No han ascendido en los últimos cinco años.

```python
func_living = df[(df['Ultima_eva']>=0.7)| (df['Tiempo_Cia']>=4) | (df['Num_proyect']>=5)]
```

```python
func_living
```

```python
correlation2 = func_living.corr()
sns.set(style= 'white' )
mask = np.zeros_like(correlation2,dtype =np.bool)
mask[np.triu_indices_from(mask)] = True

#Gráfico

f,ax = plt.subplots(figsize =(13,8))
cmap = sns.diverging_palette(10,220,as_cmap=True)

#Dibujo del heatmap de la mascara

ax = sns.heatmap(correlation2,mask=mask, cmap= cmap, vmax = 5, annot=True, annot_kws={'size':11}, square=True, 
                 xticklabels= True, yticklabels = True, linewidths = 5, cbar_kws ={'shrink':5},ax =ax)

ax.set_title('Correlacion: ¿Por qué los empleados quieren irse?',fontsize =20)
```

Resumen del análisis de datos exploratorios
- Es una empresa relativamente joven, en promedio, los empleados tienen 3 o 4 años en la empresa y los empleados de mayor edad están trabajando 10 años.
- La mayor diferencia en el salario de los que se quedaron y los que se fueron, se encontró en el departamento de administración, en los demás departamentos aunque los salarios de los que se quedaron son más altos en promedio, no es una gran diferencia.
- El número de empleados que tuvo un accidente de trabajo es de aproximadamente el 14%, de los cuales solo 169 empleados dejaron la empresa, por lo que no parece tener una correlación con la salida de los empleados.
- En cinco años solo el 2% de los empleados fueron promovidos. Es posible que muchos empleados se desmotiven y comiencen a planear irse.
- Los empleados con 7 o más en la empresa no se fueron. Los empleados con 5 años tienen más posibilidades de irse.
- Hay 2 grupos distintos de desempeño de empleados que se fueron. Un grupo con bajo desempeño con 2 proyectos y otros con alto desempeño con 5 o más proyectos. No es necesario retener a todos los empleados, el enfoque está en mantener a los empleados con un alto rendimiento.
- Los empleados con 4 años en la empresa tienen el nivel medio de satisfacción más bajo de toda la empresa con (0,47).
La satisfacción cae cuando los empleados están trabajando en 5 o más proyectos. Un número de 3 o 4 proyectos parece ser ideal independientemente del tiempo dedicado a la empresa.
- Los empleados con 5 o más proyectos que salieron también trabajaron al menos un 20% más de horas que la media de la empresa.
- El nivel de satisfacción de los empleados que se fueron se agrupa en totalmente decepcionados, por debajo de la media de satisfacción y satisfechos.
